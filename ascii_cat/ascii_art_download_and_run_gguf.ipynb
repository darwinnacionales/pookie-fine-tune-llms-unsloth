```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama

model_path = hf_hub_download(
    repo_id="darwin013/Llama-3.2-3B-ascii-cats-lora-q4_k_m-GGUF",
    filename="unsloth.Q4_K_M.gguf"
)

llm = Llama(
    model_path=model_path,
    n_ctx=2048,
    n_threads=4,
)
```

```python
import random

for i in range(3):
    print(f"\n=== Cat #{i+1} ===\n")

    prompt = random.choice([
        "ASCII cat sketch:\n",
        "Draw a cute kitten:\n",
        "ASCII cat challenge:\n",
        ""
    ])

    output = llm(
        prompt="",                    # no prompt
        max_tokens=256,
        temperature=0.8,              # 🔥 enables randomness
        top_p=0.9,
        top_k=50,
        repeat_penalty=1.1
    )
    print(output["choices"][0]["text"].strip())
```
