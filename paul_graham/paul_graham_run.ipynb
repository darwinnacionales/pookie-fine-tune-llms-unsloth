```python
from huggingface_hub import hf_hub_download

gguf_path = hf_hub_download(
    repo_id="darwin013/Meta-Llama-3.1-8B-q4_k_m-paul-graham-guide-GGUF",
    filename="unsloth.Q4_K_M.gguf"
)
```

```python
from llama_cpp import Llama

llm = Llama(
    model_path=gguf_path,
    n_ctx=2048,
    n_threads=4,  # or 6–8 for M3 if you'd like
)

prompt = "How do I build a successful startup?"
response = llm(prompt, max_tokens=256, temperature=0.8)

print("\n=== Model Response ===\n")
print(response["choices"][0]["text"])
```

```python
for chunk in llm(prompt="Write a blog post about startups.", max_tokens=256, stream=True):
    print(chunk["choices"][0]["text"], end="", flush=True)
```
